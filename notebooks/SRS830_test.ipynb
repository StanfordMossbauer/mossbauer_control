```python
import pyvisa
import time
import matplotlib.pyplot as plt 
import numpy as np
import csv 
import os
from pyvisa.constants import ControlFlow, Parity, StopBits
import pandas as pd
```

```python
gpib_address = 10
rm = pyvisa.ResourceManager()
srs830 = rm.open_resource(f"GPIB::{gpib_address}", )
```

```python
srs830.write('REST')
```

```python
# Time constants are in seconds
TIME_CONSTANTS = {
    0: 10e-6, 1: 30e-6, 2: 100e-6, 3: 300e-6, 4: 1e-3, 5: 3e-3, 6: 10e-3,
    7: 30e-3, 8: 100e-3, 9: 300e-3, 10: 1, 11: 3, 12: 10, 13: 30, 14: 100,
    15: 300, 16: 1e3, 17: 3e3, 18: 10e3, 19: 30e3
}
```

```python
def set_time_constant(instrument, time_constant):
    closest_time_constant = min(TIME_CONSTANTS.values(), key=lambda x: abs(x - time_constant))
    time_constant = list(TIME_CONSTANTS.keys())[list(TIME_CONSTANTS.values()).index(closest_time_constant)]
    instrument.write(f"OFLT {time_constant}")
    return TIME_CONSTANTS[time_constant]
```

```python
f=40
time_const = 15/f
set_time_constant(srs830, time_const)
```

```python
def read_all(instrument):
    res = instrument.query("SNAP? 3,4,9")
    R, theta, f_ref = np.array(res[:-1].split(",")).astype('float')
    return R, theta, f_ref 
```

```python
data = read_all(srs830)
```
